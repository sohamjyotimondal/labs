## Doing For tanh

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(sop):
    return np.tanh(sop)

def error(predicted, target):
    return np.power(predicted-target, 2)

def error_predicted_deriv(predicted, target):
    return 2*(predicted-target)

def tanh_sop_deriv(sop):
    return 1 - np.power(np.tanh(sop), 2)

def sop_w_deriv(x):
    return x

def update_w(w, grad, learning_rate):
    return w - learning_rate*grad

x1 = 0.1
x2 = 0.4
target = 0.7
learning_rate = 0.01
w1 = np.random.rand()
w2 = np.random.rand()
print("Initial W : ", w1, w2)
predicted_output = []
network_error = []

for k in range(80000):
    # Forward Pass
    y = w1*x1 + w2*x2
    predicted = tanh(y)
    err = error(predicted, target)
    predicted_output.append(predicted)
    network_error.append(err)
    
    # Backward Pass
    g1 = error_predicted_deriv(predicted, target)
    g2 = tanh_sop_deriv(y)

    g3w1 = sop_w_deriv(x1)
    g3w2 = sop_w_deriv(x2)

    gradw1 = g3w1*g2*g1
    gradw2 = g3w2*g2*g1
    w1 = update_w(w1, gradw1, learning_rate)
    w2 = update_w(w2, gradw2, learning_rate)

plt.figure()
plt.plot(network_error)
plt.title("Iteration Number vs Error")
plt.xlabel("Iteration Number")
plt.ylabel("Error")
plt.show()

plt.figure()
plt.plot(predicted_output)
plt.title("Iteration Number vs Prediction")
plt.xlabel("Iteration Number")
plt.ylabel("Prediction")
plt.show()
```

## RELU implementation

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(sop):
    return np.maximum(0, sop)

def error(predicted, target):
    return np.power(predicted-target, 2)

def error_predicted_deriv(predicted, target):
    return 2*(predicted-target)

def relu_sop_deriv(sop):
    return np.where(sop > 0, 1, 0)

def sop_w_deriv(x):
    return x

def update_w(w, grad, learning_rate):
    return w - learning_rate*grad

x1 = 0.1
x2 = 0.4
target = 0.7
learning_rate = 0.01
w1 = np.random.rand()
w2 = np.random.rand()
print("Initial W : ", w1, w2)
predicted_output = []
network_error = []

for k in range(80000):
    # Forward Pass
    y = w1*x1 + w2*x2
    predicted = relu(y)
    err = error(predicted, target)
    predicted_output.append(predicted)
    network_error.append(err)
    
    # Backward Pass
    g1 = error_predicted_deriv(predicted, target)
    g2 = relu_sop_deriv(y)

    g3w1 = sop_w_deriv(x1)
    g3w2 = sop_w_deriv(x2)

    gradw1 = g3w1*g2*g1
    gradw2 = g3w2*g2*g1
    w1 = update_w(w1, gradw1, learning_rate)
    w2 = update_w(w2, gradw2, learning_rate)

plt.figure()
plt.plot(network_error)
plt.title("Iteration Number vs Error")
plt.xlabel("Iteration Number")
plt.ylabel("Error")
plt.show()

plt.figure()
plt.plot(predicted_output)
plt.title("Iteration Number vs Prediction")
plt.xlabel("Iteration Number")
plt.ylabel("Prediction")
plt.show()
```

## LEAKY RELU

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(sop, alpha=0.01):
    return np.where(sop > 0, sop, alpha * sop)

def error(predicted, target):
    return np.power(predicted-target, 2)

def error_predicted_deriv(predicted, target):
    return 2*(predicted-target)

def leaky_relu_sop_deriv(sop, alpha=0.01):
    return np.where(sop > 0, 1, alpha)

def sop_w_deriv(x):
    return x

def update_w(w, grad, learning_rate):
    return w - learning_rate*grad

x1 = 0.1
x2 = 0.4
target = 0.7
learning_rate = 0.01
w1 = np.random.rand()
w2 = np.random.rand()
print("Initial W : ", w1, w2)
predicted_output = []
network_error = []

for k in range(80000):
    # Forward Pass
    y = w1*x1 + w2*x2
    predicted = leaky_relu(y)
    err = error(predicted, target)
    predicted_output.append(predicted)
    network_error.append(err)
    
    # Backward Pass
    g1 = error_predicted_deriv(predicted, target)
    g2 = leaky_relu_sop_deriv(y)

    g3w1 = sop_w_deriv(x1)
    g3w2 = sop_w_deriv(x2)

    gradw1 = g3w1*g2*g1
    gradw2 = g3w2*g2*g1
    w1 = update_w(w1, gradw1, learning_rate)
    w2 = update_w(w2, gradw2, learning_rate)

plt.figure()
plt.plot(network_error)
plt.title("Iteration Number vs Error")
plt.xlabel("Iteration Number")
plt.ylabel("Error")
plt.show()

plt.figure()
plt.plot(predicted_output)
plt.title("Iteration Number vs Prediction")
plt.xlabel("Iteration Number")
plt.ylabel("Prediction")
plt.show()
```
